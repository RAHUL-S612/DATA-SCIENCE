```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,50,100)
y=np.sin(x)
plt.plot(x,y,label="sin wave")
plt.xlabel("Rahul")
plt.ylabel("Tamilselvan")
plt.title("Friendship")
plt.legend()
plt.show()
```

```python

import numpy as np
catgories=['A','B','C','D']
values=[10,25,15,30]
plt.figure(figsize=(6,4))
plt.bar(catgories,values,color='red')
plt.xlabel('rahul')
plt.ylabel('tamil')
plt.title('barchart example' )
plt.show()
```

```python
data=(3,4,5,6,7,8,9)
plt.figure(figsize=(6,4))
plt.hist(data,bins=7,color='red',edgecolor='blue')
plt.xlabel('rahul')
plt.ylabel('tamil')
plt.title('histogram example' )
plt.show()
```
